```python
#Part 1
temp=pd.pivot_table(df, values=None, index=["Area", "Year"], columns='Variable Name')

#Part 2
temp = temp.xs('Value', axis=1, drop_level=True)

#Part 3
temp.index.rename(['Country','Year'],inplace=True)

#Part 4
temp.rename(index={"United States of America":"USA"},inplace=True)

#Part 5
temp.columns = ['GDP','NRI','PD','Area','Population']

#Part 6
df=temp
df["Area"] = df["Area"]*10 

temp.tail()
```
